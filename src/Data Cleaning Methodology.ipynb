```python
#!pip install scikit-learn
import pandas as pd

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

df = pd.read_csv("../data/labs_and_biopsies.csv", sep=",")
df.replace('Quiescent', '0', inplace=True)
df.replace('Mild', '1', inplace=True)
df.replace('Moderate', '2', inplace=True)
df.replace('Severe', '3', inplace=True)

muhcols = [
"Duo_Inflammation",
    "Gastric_Inflammation",
    "LeftColon_Inflammation",
    #"Rectum_Inflammation", # TOO MUCH DATA MISSING
    #"RightColon_Inflammation", # TOO MUCH DATA MISSING
    "TI_Inflammation",	"PGA"
    , "Hematocrit",
    "ESR",
    "CRP",
    "Albumin",
    "Vitamin D"
]
df = df[muhcols].apply(pd.to_numeric, errors="coerce")

imputer = IterativeImputer(
    estimator=RandomForestRegressor(n_estimators=100, random_state=42),
    max_iter=10,
    random_state=42
)
df_imputed = imputer.fit_transform(df)
df_complete = pd.DataFrame(df_imputed, columns=muhcols)

print(f"Shape of OG vs Shape of New: {df.shape} vs {df_complete.shape}")
print("Remaining NaNs:", df_complete.isna().sum().sum())
```

```python
df_complete["Diagnosis"] = pd.read_csv("../data/labs_and_biopsies.csv", sep=",")["Diagnosis"]

categorical_cols = [
    "Duo_Inflammation",
    "Gastric_Inflammation",
    "LeftColon_Inflammation",
    "TI_Inflammation",
    "PGA"
]

df_complete[categorical_cols] = df_complete[categorical_cols].round()
df_complete[categorical_cols] = df_complete[categorical_cols].clip(0, 3)
df_complete[categorical_cols] = df_complete[categorical_cols].astype(int) # Keeping biomarkers as floats bc BIC might be meh for the future
```

```python
print(df_complete.dtypes)
print(f"\nFinal node shape: {df_complete.shape}")
print(f"Columns: {list(df_complete.columns)}")
df_complete.to_csv("../data/final_cleaned.csv")
```
